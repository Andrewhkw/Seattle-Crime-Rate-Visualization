```python
import pandas as pd
import urllib.parse

# Initialize an empty list to store the retrieved data
data_list = []
base_url = 'https://data.seattle.gov/resource/tazs-3rd5.csv'

# Set the number of rows to retrieve in each request
limit = 50000  # Adjust this value according to your needs

# Set the initial offset
offset = 0

# Make a request loop until all the data is fetched
while True:
    # Construct the API query URL with the appropriate offset and limit
    query = f"SELECT * LIMIT {limit} OFFSET {offset}"
    encoded_query = urllib.parse.quote(query)
    query_url = f"{base_url}?$query={encoded_query}"

    # Fetch the data and append it to the data_list
    response = pd.read_csv(query_url, parse_dates=True, low_memory=False)
    data_list.append(response)

    # Check if the retrieved data has the maximum number of rows (limit)
    if len(response) < limit:
        break

    # Increment the offset for the next request
    offset += limit

# Concatenate all the retrieved data into a single DataFrame
data = pd.concat(data_list, ignore_index=True)

# Now you have the complete dataset in the 'data' variable
```

```python
data[['report_datetime', 'offense_start_datetime']] = data[['report_datetime', 'offense_start_datetime']].astype('datetime64')
```

```python
data = data.dropna(subset=['mcpp'])
```

```python
data.info()
```

```python
data['report_year'] = data['report_datetime'].dt.year
data['report_month'] = data['report_datetime'].dt.month
data['report_day'] = data['report_datetime'].dt.day
```

```python
checkoutSumByMCPP = data.groupby(['mcpp','report_year','report_month','report_day'])['offense'].count().to_frame()
```

```python
checkoutSumByMCPP = checkoutSumByMCPP.reset_index()
```

```python
mcpp_dump = ['<Null>','UNKOWN']
filtered_mcpp = np.logical_and(checkoutSumByMCPP.mcpp != '<Null>', checkoutSumByMCPP.mcpp != 'UNKWON')
```

```python
checkoutSumByMCPP= checkoutSumByMCPP[filtered_mcpp]
```

```python
date_dictionary = {'year':checkoutSumByMCPP['report_year'], 'month':checkoutSumByMCPP['report_month'],'day':checkoutSumByMCPP['report_day']}
checkoutSumByMCPP['date'] = pd.to_datetime(date_dictionary)
```

```python
# Define dropdown selection for species
mcpp_dropdown = alt.binding_select(options=sorted(list(checkoutSumByMCPP.mcpp.unique())))
mcpp_select = alt.selection_single(fields=['mcpp'], bind=mcpp_dropdown, name='Area (MCPP)')

# Scatter plot of petal length vs. petal width
offense_plot = alt.Chart(checkoutSumByMCPP).mark_line().encode(
    alt.X('date:T', title='Offense Date'),
    alt.Y('sum(offense):Q', title='Offense Count'),
    #alt.Color('UsageClass:N', title='Usage Class'),
    #tooltip=['offense:N', 'crime_against_category:N', 'offense_parent_group:N']
).add_selection(
    mcpp_select
).transform_filter(
    mcpp_select
).properties(
    width=600,
    height=400
)

# Combine the scatter plot and histogram into a single chart
chart = alt.hconcat(offense_plot)

chart.properties(title='Offense Dataset by Area (MCPP)')
```
